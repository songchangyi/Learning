#### Resampling

```python
rng = pd.date_range('1/1/2011', periods=90, freq='d')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts.head()
```

```python
type(ts)
```

```python
ts.resample('M').sum()
```

```python
ts.resample('3d').sum()
```

```python
rng = pd.date_range('1/1/2011', periods=90, freq='3d')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
```

```python
ts.resample('d').asfreq().head()
```

```python
ts.resample('d').ffill(1)
```

```python
ts.resample('d').interpolate('linear')
```

#### Slide window

```python
rng = pd.date_range('7/1/2016', periods=600, freq='d')
df = pd.Series(np.random.randn(len(rng)), index=rng)
df.head()
```

```python
r = df.rolling(window=10)
r
```

```python
r.mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15, 5))

df.plot(style='r--')
df.rolling(window=10).mean().plot(style='b')
```
